```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/student_data.csv")
df.head()
```

```python
df.info()

```

```python
df.describe()
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
sns.histplot(df['G3'], kde=True)
plt.title("Grades Distribution")
plt.show()
```

```python
sns.boxplot(x='sex', y='G3', data=df)
plt.title("Gender vs Grades")
plt.show()
```

```python
sns.scatterplot(x='health', y='G3', data=df)
plt.title("Health vs Grades")
plt.show()
```

```python
sns.boxplot(x='famrel', y='G3', data=df)
plt.title("Family Relation vs Grades")
plt.show()
```

**Most students scored in the average range**

Students with better health tend to perform better

Strong family relationships may positively affect grades

Parental education appears to influence performance

Performance varies across genders
